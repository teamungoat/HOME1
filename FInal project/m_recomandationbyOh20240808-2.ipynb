```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from konlpy.tag import Okt
```

```python
# 한국어 텍스트 처리를 위한 Okt 초기화
okt = Okt()
```

```python
# 텍스트를 토큰화하고 원형으로 복원하는 함수 정의
def tokenize_and_lemmatize(text):
    if isinstance(text, str):  # Check if 'text' is a string
        tokens = okt.morphs(text, stem=True)
        return ' '.join(tokens)
    else:
        return ''  # Return an empty string or handle NaN case appropriately
```

```python
# 텍스트 데이터에 함수 적용
data['Processed_Text'] = data['Text'].apply(tokenize_and_lemmatize)
```

```python
# 처리된 텍스트 데이터 벡터화
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['Processed_Text'])
```

```python
import spacy
import re

# CSV 파일을 Pandas DataFrame으로 불러오기
file_path = '/content/movie_data'
data = pd.read_csv(file_path)

# 'Unnamed: 0' 열 제거
data = data.drop(columns=['Unnamed: 0'])

# 열 이름을 한국어로 설정
data.columns = ['영화명', '평점', '연도', '상영시간', '연령', '감독', '출연진', '줄거리', '장르']

pd.set_option('display.max_columns', None)

# spaCy 한국어 모델 로드
nlp = spacy.load("ko_core_news_sm")

# 한국어 불용어 정의
stop_words_ko = {"및", "도", "에", "의", "가", "이", "은", "는", "을", "를", "에서", "로", "과", "와", "한", "그", "이", "하", "여", "속", "전", "자", "이다"}

# 장르 동의어 매핑 생성
genre_synonyms_ko = {
    '공포': ['공포', '호러', '스릴러', '무서운'],
    '로맨스': ['로맨스', '사랑', '연애', '로맨틱'],
    '멜로': ['멜로', '멜로드라마'],
    '액션': ['액션', '모험', '전투', '스릴'],
    '코미디': ['코미디', '웃긴', '유머', '희극'],
    '서부': ['서부', '웨스턴', '카우보이', '서부극', 'western'],
}
```

```python
# 장르 동의어를 정규화하는 함수
def normalize_genre_from_synonyms_ko(keyword):
    keyword = keyword.strip().lower()
    for genre, synonyms in genre_synonyms_ko.items():
        if keyword in [synonym.lower() for synonym in synonyms]:
            return genre
    return keyword

# 장르 정규화 함수 (괄호 및 슬래시 처리 포함)
def normalize_genre(text):
    if pd.isna(text):
        return ''
    text = re.sub(r'\s*\(/?\s*|\s*/\s*|\s*\)\s*', ',', text)
    text = re.sub(r'\s+', ' ', text)
    genres = [normalize_genre_from_synonyms_ko(genre.strip()) for genre in text.split(',')]
    return ','.join(sorted(set(genres)))

# 문장에서 키워드 추출하는 함수
def extract_keywords(sentence):
    doc = nlp(sentence)
    keywords = [token.text.strip() for token in doc if token.text.strip() not in stop_words_ko and token.pos_ in ['NOUN', 'ADJ', 'PROPN']]


    # 불용어를 제거한 키워드 정리
    cleaned_keywords = []
    for keyword in keywords:
        # 공백으로 나누고, 불용어 제거
        parts = re.split(r'\s+', keyword)
        cleaned_parts = [part for part in parts if part not in stop_words_ko]
        cleaned_keyword = ' '.join(cleaned_parts)
        cleaned_keywords.append(cleaned_keyword)

    return cleaned_keywords

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from konlpy.tag import Okt
import re

# 데이터 로드

movie_titles = data['영화명'].dropna().tolist()

# TF-IDF 벡터화
tfidf = TfidfVectorizer()
tfidf_movie_names = tfidf.fit_transform(movie_titles)

# 사용자 입력 토큰화 및 벡터화
def preprocess_and_vectorize(user_input, tfidf_vectorizer):
    okt = Okt()
    # 텍스트 전처리 및 토큰화
    user_input_text = re.sub("[^가-힣ㄱ-ㅎㅏ-ㅣ0-9\\s]", "", user_input)
    tokens = okt.morphs(user_input_text, stem=True)

    # 각 토큰을 TF-IDF 벡터로 변환
    token_vectors = tfidf_vectorizer.transform(tokens)
    return tokens, token_vectors

def get_similarity_scores(tokens, token_vectors, tfidf_matrix):
    # 각 토큰 벡터와 데이터의 영화 제목 벡터 간의 유사도 계산
    similarity_scores = {}
    for token, token_vector in zip(tokens, token_vectors):
        token_similarities = cosine_similarity(token_vector, tfidf_matrix)
        # 유사도와 영화 제목을 함께 저장
        similarity_scores[token] = [(movie_titles[i], similarity) for i, similarity in enumerate(token_similarities.flatten())]
    return similarity_scores

def find_most_similar_movies(user_input, tfidf_vectorizer, tfidf_matrix, movie_titles):
    tokens, token_vectors = preprocess_and_vectorize(user_input, tfidf_vectorizer)
    similarity_scores = get_similarity_scores(tokens, token_vectors, tfidf_matrix)

    # 각 토큰별로 유사도가 가장 높은 영화 제목을 선택
    most_similar_movies = {}
    for token, scores in similarity_scores.items():
        # 유사도가 가장 높은 영화 제목을 선택
        most_similar_movies[token] = max(scores, key=lambda x: x[1])

    return most_similar_movies

# 사용자 입력 예시
user_input = "극장판짱구는못말려 같은 재미있는 영화"  # 사용자 입력을 적절히 수정하세요

# 유사한 영화 제목 찾기
most_similar_movies = find_most_similar_movies(user_input, tfidf, tfidf_movie_names, movie_titles)

# 결과 출력
for token, (movie_title, similarity) in most_similar_movies.items():
    print(f"토큰: {token}, 가장 유사한 영화 제목: {movie_title}, 유사도: {similarity:.4f}")
```
