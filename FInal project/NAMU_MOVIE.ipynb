```python
# '원작' 항목이 '출연'이고, 그 다음 항목이 실제 출연 정보인지 확인하여 재배치
fixed_movie_list1 = []
skip_next = False
for i in range(len(movie_data)):
    movie_data_dict = movie_data[i][1]
    movie_data_df = pd.DataFrame(movie_data_dict).reset_index()
    movie_data_df.columns = ['kind', 'content']# 열 이름 변경
    new_rows = []
    for idx in range(len(movie_data_df)):
        if skip_next:
            skip_next = False
            continue

        if movie_data_df.loc[idx, 'kind'] == '원작' and movie_data_df.loc[idx, 'content'] == '출연':
            new_rows.append(['원작', ''])
            new_rows.append(['출연', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '편집':
            new_rows.append(['편집', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '미술':
            new_rows.append(['미술', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '촬영':
            new_rows.append(['촬영', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '상영 시간':
            new_rows.append(['상영 시간', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '제작비':
            new_rows.append(['제작비', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '월드 박스오피스':
            new_rows.append(['월드 박스오피스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '북미 박스오피스':
            new_rows.append(['북미 박스오피스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '대한민국 총 관객 수':
            new_rows.append(['대한민국 총 관객 수', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '상영 등급':
            new_rows.append(['상영 등급', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        elif movie_data_df.loc[idx, 'content'] == '시놉시스':
            new_rows.append(['시놉시스', movie_data_df.loc[idx + 1, 'kind']])
            skip_next = False
        else:
            new_rows.append([movie_data_df.loc[idx, 'kind'], movie_data_df.loc[idx, 'content']])
    fixed_movie_list1.append(new_rows)
```

```python
file_path = '/content/fixed_movie_data'

def load_data(file_path):
    # 데이터를 저장할 변수
    movies = {}
    current_index = None
    current_data = []

    with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split(',')
            index = parts[0].strip()
            data = ','.join(parts[1:]).strip()

            if index.isdigit() and len(parts) > 1:
                if parts[1].strip() == 'movie name':
                    if current_index is not None:
                        movies[current_index] = '\n'.join(current_data)
                    current_index = index
                    current_data = [line.strip()]
                else:
                    current_data.append(line.strip())
            else:
                if current_index is not None:
                    current_data.append(line.strip())

        # 마지막 영화 데이터 추가
        if current_index is not None:
            movies[current_index] = '\n'.join(current_data)

    return movies

def find_movie_index(movies, movie_title):
    for index, data in movies.items():
        if f"movie name,{movie_title}" in data:
            return index
    return None

def print_movie_data(movies, index):
    if index in movies:
        print(f"영화 데이터 (인덱스 {index}):\n{movies[index]}")
    else:
        print(f"인덱스 {index}의 데이터를 찾을 수 없습니다.")

# 데이터 로딩
movies = load_data(file_path)

# 사용자로부터 영화 제목 입력받기
movie_title = input("영화 제목을 입력하세요: ").strip()

# 영화 제목에 해당하는 인덱스 찾기
index = find_movie_index(movies, movie_title)

# 해당 인덱스를 기준으로 데이터 출력
if index is not None:
    print_movie_data(movies, index)
else:
    print(f"영화 '{movie_title}'를 찾을 수 없습니다.")
```

```python
import pandas as pd

# 데이터 로드
df = pd.read_csv('/content/fixed_movie_data')

# 영화 정보를 저장할 딕셔너리 생성
movie_info = {}

# 포함할 정보의 종류
relevant_kinds = ['영화명', '장르', '감독', '출연', '배급', '개봉일', '상영시간', '관람등급', '시놉시스']

# 데이터프레임을 반복하여 딕셔너리 채우기
for i, row in df.iterrows():
    movie_id = row['movie_id']
    kind = row['kind']
    content = str(row['content'])  # 모든 값을 문자열로 변환

    # 관련 없는 정보는 건너뛰기
    if kind not in relevant_kinds:
        continue

    if movie_id not in movie_info:
        movie_info[movie_id] = {}

    movie_info[movie_id][kind] = content

# 영화 제목과 정보를 매핑하는 새로운 데이터프레임 생성
movie_titles = df[df['kind'] == '영화명'][['movie_id', 'content']].rename(columns={'content': 'title'})
movie_titles['info'] = movie_titles['movie_id'].apply(lambda x: {k: v for k, v in movie_info[x].items() if k in relevant_kinds})

# 불용어 리스트 (필요에 따라 추가 가능)
stop_words = ['의', '이', '가', '은', '는', '을', '를', '에', '와', '과', '도', '로', '에서', '해줘']

# 불용어를 제거하는 함수
def remove_stop_words(text):
    words = str(text).split()
    filtered_words = [word for word in words if word not in stop_words]
    return ' '.join(filtered_words)

# 제목으로 영화 정보를 가져오는 함수
def get_movie_info(title):
    try:
        filtered_title = remove_stop_words(title)
        movie_id = movie_titles[movie_titles['title'].str.contains(filtered_title, case=False, na=False)].iloc[0]['movie_id']
        return movie_titles[movie_titles['movie_id'] == movie_id]['info'].iloc[0]
    except IndexError:
        return f"영화를 찾지 못했습니다.: {title}"

# 장르별로 영화를 추천하는 함수
def recommend_movies_by_genre(genre):
    recommended_movies = movie_titles[movie_titles['info'].apply(lambda x: '장르' in x and genre in x['장르'])]
    return recommended_movies[['title', 'info']]

# 장르 목록을 가져오는 함수
def get_genres():
    genres = set()
    for info in movie_titles['info']:
        if '장르' in info:
            genres.update(info['장르'].split())
    return sorted(genres)

# 사용자 입력 처리
user_input = input("영화 제목이나 원하는 장르를 입력하세요 (예: 공포, 액션, 드라마): ")
processed_input = remove_stop_words(user_input)

# 장르 목록을 가져와서 사용자 입력이 장르인지 제목인지 확인
genres = get_genres()
genre_found = None
for genre in genres:
    if genre in processed_input:
        genre_found = genre
        break

if genre_found:
    # 사용자가 장르를 입력했을 때 추천 영화 출력
    recommended_movies = recommend_movies_by_genre(genre_found)
    if not recommended_movies.empty:
        print(f"추천 {genre_found} 영화 목록:")
        for index, row in recommended_movies.iterrows():
            for kind, content in row['info'].items():
                print(f"{kind}: {content}")
            print("-" * 50)
    else:
        print(f"추천할 {genre_found} 영화가 없습니다.")
else:
    # 사용자가 제목을 입력했을 때 영화 정보 출력
    movie_info = get_movie_info(user_input)
    print(movie_info)
```
