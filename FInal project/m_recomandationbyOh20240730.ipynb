```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel
from konlpy.tag import Okt
```

```python
data['연도'] = data['연도'].astype(str)
data['평점'] = data['평점'].astype(str)
```

```python
# NLP를 위한 텍스트 특성 결합
# Adjusted column names to match the new names
data['Text'] = (data['영화명'] + ' ' + data['평점'] + ' ' + data['연도'] + ' ' + data['상영시간'] + ' ' + data['연령']  + ' ' + data['감독'] + ' ' + data['출연']+ ' ' + data['시놉시스'])
```

```python
# 한국어 텍스트 처리를 위한 Okt 초기화
okt = Okt()
```

```python
# 텍스트를 토큰화하고 원형으로 복원하는 함수 정의
def tokenize_and_lemmatize(text):
    if isinstance(text, str):  # Check if 'text' is a string
        tokens = okt.morphs(text, stem=True)
        return ' '.join(tokens)
    else:
        return ''  # Return an empty string or handle NaN case appropriately
```

```python
# 텍스트 데이터에 함수 적용
data['Processed_Text'] = data['Text'].apply(tokenize_and_lemmatize)
```

```python
# 처리된 텍스트 데이터 벡터화
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(data['Processed_Text'])
```

```python
# 코사인 유사도 행렬 계산
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
def get_recommendations(title, cosine_sim=cosine_sim):
    if title not in data['영화명'].values:
        return f"영화명 '{title}'을(를) 데이터셋에서 찾을 수 없습니다."

    idx = data[data['영화명'] == title].index
    if len(idx) == 0:
        return f"영화명 '{title}'을(를) 데이터셋에서 찾을 수 없습니다."

    idx = idx[0]  # Take the first index if multiple entries are found (though unlikely)

    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    sim_scores = sim_scores[1:11]  # 상위 10개
    movie_indices = [i[0] for i in sim_scores]
    return data['영화명'].iloc[movie_indices].tolist()
```

```python
def get_recommendations(title=None, movie_rate=None, director=None, year=None, actor=None, cosine_sim=cosine_sim):
    if title:
        if title not in data['영화명'].values:
            return f"영화명 '{title}'을(를) 데이터셋에서 찾을 수 없습니다."
        idx = data[data['영화명'] == title].index
        if len(idx) == 0:
            return f"영화명 '{title}'을(를) 데이터셋에서 찾을 수 없습니다."
        idx = idx[0]  # Take the first index if multiple entries are found (though unlikely)
        sim_scores = list(enumerate(cosine_sim[idx]))
        sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
        sim_scores = sim_scores[1:11]  # 상위 10개 추천
        movie_indices = [i[0] for i in sim_scores]
        recommendations = data.iloc[movie_indices]
    else:
        conditions = []
        if movie_rate:
            # 평점을 문자열에서 float으로 변환하여 비교
            movie_rate = float(movie_rate)
            conditions.append(data['평점'].astype(float) >= movie_rate)
        if director:
            conditions.append(data['감독'].str.contains(director, na=False))
            print(conditions)
        if year:
            conditions.append(data['연도'].astype(str).str.contains(year, na=False))
        if actor:
            conditions.append(data['출연'].str.contains(actor, na=False))
        if not conditions:
            return "영화명, 평점, 감독 또는 연도 중 하나를 입력하세요."
        recommendations = data
        for condition in conditions:
            recommendations = recommendations[condition]
        recommendations = recommendations.head(10)

    recommendations = recommendations.copy()
    # recommendations.loc[:, '연도'] = recommendations['연도'].astype(int)
    return recommendations[['영화명', '평점', '감독', '연도']].values.tolist()

# 사용자 입력을 통한 영화명, 평점, 감독, 제작 연도 입력
movie_title = input("추천을 받고 싶은 영화명을 입력하세요 (생략 가능): ").strip()
movie_rate = input("추천을 받고 싶은 평점를 입력하세요 (생략 가능): ").strip()
director = input("추천을 받고 싶은 감독을 입력하세요 (생략 가능): ").strip()
year = input("추천을 받고 싶은 제작 연도를 입력하세요 (생략 가능): ").strip()
actor = input("추천을 받고 싶은 배우를 입력하세요 (생략 가능): ").strip()
# 빈 문자열을 None으로 변환
movie_title = movie_title if movie_title else None
movie_rate = movie_rate if movie_rate else None
director = director if director else None
year = year if year else None
actor = actor if actor else None

recommended_movies = get_recommendations(movie_title, movie_rate, director, year, actor)
print("추천된 영화들:\n", recommended_movies)
```
