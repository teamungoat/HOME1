##1. 데이터 가져오기

```python
# proverb.txt
contents=""" 인공지능과 빅데이터는 현대 사회에서 중요한 기술이다.
많은 회사들이 데이터 분석을 통해 비즈니스 전략을 수립하고 있다.
데이터 과학과 기계 학습은 이러한 분석에서 핵심적인 역할을 한다.
최근에는 딥러닝과 같은 고급 기술이 많은 관심을 받고 있다.
이 외에도 클라우드 컴퓨팅, 사물 인터넷, 자동화 기술 등이
기업과 일상 생활에 영향을 미치고 있다.
특히, 프로그래밍 언어인 파이썬은 인공지능 연구와 개발에서 널리 사용된다.
또한, 영어 학습과 같은 교육 분야에서도 AI 기술이 점점 더 활용되고 있다."""
file=open("/content/1.03.wordcloud.txt", 'w') #mode w
file.write(contents)
file.close()
# hello.txt
file=open("./hello.txt",'a')
while 1:
  content=input()
  if not content: break
  file.write(content+ '\n')
  print(content)
file.close()

# fruits.txt
fruits=["apple","apple","apple","banana","banana","orange","mango","orange","orange","orange"]
index=len(fruits)
file=open("./fruits.txt",'a')
for i in range(len(fruits)):
  if i < (index-1):
    file.write(fruits[i]+'\t')
  else:
    file.write(fruits[i]+'\n')
file.close()
```

```python
import os
import glob
f1=[]
for file in glob.glob('*.txt'):
  with open(file, 'r', newline='') as filereader:
    for row in filereader:
      contents=row.split()
      for i in contents:
        contents=i.strip('.').strip(',').strip('')
        f1.append(contents)
f1
```

```python
import os
import glob

f1=[]
for file in glob.glob('*.txt'):
  with open(file, 'r', newline='') as filereader:
    for row in filereader:
      contents=row.split()
      for i in contents:
        contents=i.strip('.').strip(',').strip('').strip('를').strip('을').strip("''")
        contents=contents.strip('합니다').strip('하세요')
        contents=contents.strip('에')
        f1.append(contents)
f1
```

##2. 단어의 빈도수 구하기

```python
from collections import Counter
f1
word_counts=Counter(f1)
dict(word_counts)
# word_counts.items()
wc=sorted(word_counts.items(),key=lambda count:count[1], reverse=True)
wc
```

```python
import pandas as pd
wcdf=pd.DataFrame(wc)
wcdf.columns=['word', 'freq']
```

```python
wcdf1=wcdf.set_index('word')
wcdf1.head()
```

```python
wordcloud=WordCloud(font_path='NanumGothic.ttf', background_color='white', colormap
='autumn').generate_from_frequencies(word_freq)
plt.imshow(wordcloud)
plt.title('Word Frequency', size=13) # 제목 지정
plt.show()
```

```python
from PIL import Image
import numpy as np
im=Image.open('heart-26-1.png') # 이미지 파일 읽어오기
mask_arr=255-np.array(im) # 픽셀 값 배열 형태 변환
wordcloud=WordCloud(font_path='NanumGothic.ttf', background_color='white', colormap='autumn',width=700,height=700,random_state=43,mask=mask_arr,prefer_horizontal=True).generate_from_frequencies(word_freq)
plt.figure(figsize=(6, 6))
plt.imshow(wordcloud)
plt.title('Word Frequency', size=13)
plt.axis('off')
plt.show()
```
