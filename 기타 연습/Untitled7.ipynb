```python
text = "아이엠어보이유알어걸"

def print_5xn(string):
    for i, t in enumerate(string):
        if(i % 5 == 4):
            print(t)
        else:
            print(t, end='')
print_5xn(text)
```

```python
def func_db(res):
#변경할 수식
    return str(int(res.group())*2)

re.sub("\d+",func_db,"10 20 sky 50 100 bird 200")
```
