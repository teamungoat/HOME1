```python
#21 입력된 문자열을 역순으로 출력하는 print_reverse 함수를 정의하라.
def print_reverse(string):
    print(string[::-1])

string = input("문자열 입력: ")
print_reverse(string)
```

```python
#22 성적 리스트를 입력 받아 평균을 출력하는 print_score 함수를 정의하라.
def print_score_averge(score1, score2, score3):
    print((score1+score2+score3)/3)

score_input1, score_input2, score_input3 = map(int,input("점수를 입력하시오: ").split())
print_score_averge(score_input1, score_input2, score_input3)

```

```python
#23 하나의 리스트를 입력받아 짝수만 화면에 출력하는 print_even 함수를 정의하라.
num_list_input = input("숫자를 입력하시오: ").split()
numbers = list(map(int, num_list_input))
def print_even(numbers):
    for number in numbers:
        if number % 2 == 0:
            print(number)
print_even(numbers)
```

```python
#24 하나의 딕셔너리를 입력받아 딕셔너리의 key 값을 화면에 출력하는 print_keys 함수를 정의하라.
def print_keys(my_dict):
    for key in my_dict.keys():
        print(key)

print_keys ({"이름":"김말똥", "나이":30, "성별":0})
```

```python
#25 my_dict와 날짜 키값을 입력받아 OHLC 리스트를 출력하는 print_value_by_key 함수를 정의하라.
my_dict = {"10/26" : [100, 130, 100, 100],
           "10/27" : [10, 12, 10, 11]}

def print_value_by_key(my_dict, key):
    print(my_dict[key])

print_value_by_key(my_dict, "10/26")
```

```python
#26 입력 문자열을 한 줄에 다섯글자씩 출력하는 print_5xn(string) 함수를 작성하라.
def print_5xn(string):
    for i in range(0, len(string), 5):
        print(string[i:i+5])
print_5xn("아이엠어보이유알어걸")

```

```python
#27 문자열과 한줄에 출력될 글자 수를 입력을 받아 한 줄에 입력된 글자 수만큼 출력하는 print_mxn(string) 함수를 작성하라.
def print_mxn(string, length_of_string):
    for i in range(0, len(string), length_of_string):
        print(string[i:i+length_of_string])

print_mxn("아이엠어보이유알어걸", 3)
```

```python
#28 연봉을 입력받아 월급을 계산하는 calc_monthly_salary(annual_salary) 함수를 정의하라. 회사는 연봉을 12개월로 나누어 분할 지급하며, 이 때 1원 미만은 버림한다.
def calc_monthly_salary(annual_salary):
    monthly_salary = annual_salary/12
    print(int(monthly_salary))

calc_monthly_salary(12000000)
```

```python
#34 숫자로 구성된 하나의 리스트를 입력받아, 짝수들을 추출하여 리스트로 반환하는 pickup_even 함수를 구현하라.
even_number = []
def pickup_even(num_list):
    for i in num_list:
        if i %2 ==0:
            even_number.append(i)
    print(even_number)
pickup_even([3, 4, 5, 6, 7, 8])
```
