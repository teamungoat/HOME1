```python
#19
def print_arithmetic_operation(a, b):
    print(a+b)
    print(a-b)
    print(a*b)
    print(a/b)
num1, num2 = map(int, input("두개의 숫자를 입력하시오: ").split())
print_arithmetic_operation(num1, num2)
```

```python
#20
def max_number(a, b, c):
    if a > b and a > c:
        print("max_number", a)
    elif b > a and b > c:
        print("max_number", b)
    else:
        print("max_number", c)

num1, num2, num3 = map(int, input("숫자 3개를 입력").split())
max_number(num1, num2, num3)

```
