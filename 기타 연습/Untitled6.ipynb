```python
# 7  DashInsert 함수
# DashInsert 함수는 숫자로 구성된 문자열을 입력받은 뒤 문자열 안에서 홀수가 연속되면 두 수 사이에 - 를 추가하고,
# 짝수가 연속되면 * 를 추가하는 기능을 갖고 있다. DashInsert 함수를 완성하시오.

# 입력 예시:4546793
# 출력 예시: 454*67-9-3
```

```python
def DashInsert():
  s=str(input('숫자입력'))
  res=''
  for i in range(len(s)-1):
    if int(s[i])%2==1 and int(s[i+1])%2==1:   #연속된숫자가 홀수이면
      res=res+s[i]+'-'
    elif int(s[i])%2==0 and int(s[i+1])%2==0:   #연속된숫자가 짝수이면
      res=res+s[i]+'*'
    else:
      res=res+s[i]
  return res+s[-1]  #입력된 마지막 숫자를 추가

DashInsert()
```

```python

def dashinsert():
    s = str(input('숫자를 입력하시오'))
    res = ' '
    for i in range(len(s)-1):
        if int(s[i])%2==1 and int(s[i+1])%2==1:
            res = res+(s[i])+'-'
        elif int(s[i])%2==0 and int(s[i+1])%2==0:
            res=res+s[i]+'*'
        else:
            res=res+s[i]
    return res+s[-1]
dashinsert()

```

```python
#8
# 문자열을 입력받아 같은 문자가 연속적으로 반복되는 경우에 그 반복 횟수를 표시해 문자열을 압축하여 표시하시오.
# 입력 예시: aaabbcccccca
# 출력 예시: a3b2c6a1
```

```python
s = input('문자열을 입력하시오')
def compression():
    result = ' '
    count = 1
    for i in range(len(s)):
        if result == ' ':
            result += s[i]
        elif result[-1] == s[i]:
            count+=1
        else:
            result = result + str(count)+s[i]
            count = 1
        if i == len(s)-1:
            result = result + str(count)
    return result
compression()



```
