```python
num1, num2 = map(int, input().split())
sum = 0
for i in range(num1, num2+1):
    text = ''
    for j in str(i):
        if(j != '0'):
            text += j+'*'
        else:
            text += '0*'
    print(text[:3])
    sum += eval(text[:3])  #1*1*
print(sum)
```
