```python
def hi(cnt): #cnt = 0
    if cnt==0:
        return #함수를 호출한 곳으로 되돌아가라
    print("(before)hi", cnt)
    cnt-=1
    hi(cnt)
    print("(after)hi", cnt)
"""
hi(3) -> hi(2) -> hi(1) -> hi(0)
"""
```

```python
hi(3)
```

```python
class Person:
    def hello(self): #객체를 생성한 다음, 객체 변수를 통해 호출
        print("hello 함수를 호출했습니다")
        print("저의 이름은 {0}이며, 나이는 {1}입니다".format(self.name, self.age))
    def __init__(self,na,ag,addr): #객체가 생성되는 시점에 자동 호출
        self.mymsg="반갑습니다"
        print("잘가세요")
        self.name=na
        self.age=ag
        self.address=addr
```

```python
hgd.age
```

```python
hgd=Person("홍길동", 25, "서울시")
```

```python
hgd.hello()
```

```python
p=Person()
p.hello()
```

```python
class Animal:
    def __init__(self,nl,ag,sp):
        self.num_leg=nl
        self.age=ag
        self.species=sp
        print("객체 초기화 함수(init)가 호출되었습니다")
    def sleep(self):
        print("{0} 종류의 동물이며 {1}개의 다리를 가지고 있습니다".format(self.species, self.num_leg))
```

```python
puppy=Animal(4, 20, "Mammalia")
```

```python
puppy.sleep()
```

```python
baduki=Animal(4, 20, "Mammalia")
```

```python
baduki.sleep()
```
