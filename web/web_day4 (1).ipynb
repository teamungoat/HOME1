```python
url="https://www.melon.com/chart/index.htm"
driver.get(url)
```

```python
soup=BeautifulSoup(html, "html.parser")
soup
```

```python
#len(soup.select("#conts_section"))
soup.select("#conts_section")
```

```python
songs=soup.select("tr")
songs[100]
```

```python
songs=songs[1:]
songs
```

```python
song=songs[0]
```

```python
song
```

```python
song.select("div.ellipsis.rank01 > span > a")[0].string
song.select("div.ellipsis.rank01 > span > a")[0].text
```

```python
song.select("div.ellipsis.rank02 > span > a")[0].string
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select("div.ellipsis.rank01 > span > a")[0].text
    singer=song.select("div.ellipsis.rank02 > span > a")[0].text
    song_data.append(['Melon',rank, title, singer]) #  [[  , ,  ], [  ], ... ,[  ]]
    print(rank, title, singer, sep=" | ")
    rank+=1    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(song_data, columns=['ÏÑúÎπÑÏä§', 'ÏàúÏúÑ', 'Ï†úÎ™©', 'Í∞ÄÏàòÎ™Ö'])
df.to_excel("melon.xlsx", index=False)
```
