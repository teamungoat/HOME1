```python
url="https://www.melon.com/chart/index.htm"
driver.get(url)
```

```python
soup=BeautifulSoup(html, "html.parser")
soup
```

```python
#len(soup.select("#conts_section"))
soup.select("#conts_section")
```

```python
songs=soup.select("tr")
songs[100]
```

```python
songs=songs[1:]
songs
```

```python
song=songs[0]
```

```python
song
```

```python
song.select("div.ellipsis.rank01 > span > a")[0].string
song.select("div.ellipsis.rank01 > span > a")[0].text
```

```python
song.select("div.ellipsis.rank02 > span > a")[0].string
```

```python
song_data=[]
rank=1
for song in songs:
    title=song.select("div.ellipsis.rank01 > span > a")[0].text
    singer=song.select("div.ellipsis.rank02 > span > a")[0].text
    song_data.append(['Melon',rank, title, singer]) #  [[  , ,  ], [  ], ... ,[  ]]
    print(rank, title, singer, sep=" | ")
    rank+=1    
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(song_data, columns=['서비스', '순위', '제목', '가수명'])
df.to_excel("melon.xlsx", index=False)
```

```python
"""
뉴스기사 수집 : '환율' 단어가 포함된 기사

# [xxx 기자]
# "오늘의 달러 환율은... 영향을 받아 ... 환열..."
# 손흥민 선수가 골을 넣어습니다....

# """

# 1) 뉴스 기사 스크래핑 -> 임시 저장
# 2) 임시 저장된 텍스트에서 '환율' 단어 유무 확인(?) -> 정규식(search, find_all)

news = "오늘의 달러 환율은... 영향을 받아 ... 환율..."
```

```python
import re
if re.search('환율', news):
    #db에 저장
    print("저장해야 할 뉴스기사")
else:
    print("버릴 뉴스기사")
```

```python
#최소 1개 이상 구현
1) 벅스뮤직(스크래핑 1위~100위)
2) 유튜브 랭크(1위~5000위) 추출 - 유튜브방송국이름, 유튜버, 가입자수
3) 보배드림 '국산차' 검색 - 1페이지~10페이지까지, 페이지당 50개씩
4) 인스타그램 가입, 스타벅스(내일)
```
