```python
m1=np.array([0,1,1])
m2=np.array([1,0,1])
m3=np.array([2,2,2])
m4=np.array([1,1,1])
```

```python
from numpy.linalg import norm
```

```python
def cossim(a,b):
    return np.dot(a,b)/(norm(a)*norm(b)) #내적/놀음
```

```python
print(cossim(m1,m4))
print(cossim(m2,m4))
print(cossim(m3,m4))
```
