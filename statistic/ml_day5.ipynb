```python
data = pd.read_csv("snsdata.csv")

null_data=data[data['gender'].isna()] #null data추출
null_feature=null_data[null_data.columns[4:40]]

non_null_data=data[data['gender'].notna()] #트레이닝 용도로 gender에서 null이 아닌 데이터 추출
non_null_data

nonnull_feature=non_null_data[non_null_data.columns[4:40]] #feature 추출
nonnull_feature

target=non_null_data[non_null_data.columns[1]] #notnull data에서 젠더열 추출
target

def gender_incoding(x): #target 열을 int로 mapping
    if x == 'M':
        return 1
    elif x == 'F':
        return 2

mapped_target=target.apply(gender_incoding)
mapped_target

from sklearn.neighbors import KNeighborsClassifier #KNeightborsClassifier로 남자인지 여자인지 분류
from sklearn.model_selection import KFold
K_fold=KFold(n_splits=10, shuffle=True, random_state=True)

clf=KNeighborsClassifier(n_neighbors=5)

clf.fit(nonnull_feature,mapped_target)

null_data['gender']=clf.predict(null_feature) #null_data의 젠더열에 예측한 결과를 데입
null_data

data.loc[data['gender'].isna(), 'gender'] = null_data['gender'] #본래의 data에 예측한 성별을 데입
```

```python
data['age'].fillna(data.groupby('gradyear')['age'].transform('mean'), inplace=True) #age 의 nan값에 열도별로 구릅화한 후 age의 평균값을 데입
```

```python
data.info()
```

```python
from sklearn.model_selection import cross_val_score

for i in range(1,20):  #아주 오래걸림
    clf=KNeighborsClassifier(n_neighbors = i)
    score=cross_val_score(clf,nonnull_feature,mapped_target,cv=K_fold, scoring = 'accuracy')
    print(f'{i} 확률은 {round(np.mean(score),3)}')
```
