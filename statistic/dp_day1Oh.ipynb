```python
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df=pd.read_csv("GroceryStoreDataSet.csv", names=['products'], sep=',')
df
```

```python
data=list(df['products'].apply(lambda x:x.split(",")))
data
```

```python
from mlxtend.preprocessing import TransactionEncoder
```

```python
te=TransactionEncoder()
```

```python
trans_data=te.fit(data).transform(data)
```

```python
trans_data
```

```python
te.columns_
```

```python
df=pd.DataFrame(trans_data, columns=te.columns_)
df
```

```python
df=apriori(df, min_support=0.2, use_colnames=True)
df
```

```python
df_ar=association_rules(df, metric="confidence", min_threshold=0.6)
df_ar
```

```python
data=pd.read_csv('chipotle.csv')
data
```

```python
order_item_df = data.groupby('order_id')['item_name'].apply(list).reset_index()
order_item_df
```

```python
def show_unique_items(dataframe, column_name):
    unique_items = dataframe[column_name].explode().unique()
    return unique_items
```

```python
unique_item_names = show_unique_items(order_item_df, 'item_name')
unique_item_names
```

```python
def flatten_list(cell):
    if isinstance(cell, list):
        return cell[0:]  # 여기서 [0]은 리스트의 첫 번째 요소를 가져오는 예시입니다. 필요에 따라 수정 가능합니다.
    else:
        return cell
```

```python
order_item_df['item_name']=order_item_df.item_name.apply(flatten_list)
order_item_df
```

```python
list_item_name=list(order_item_df['item_name'].apply(lambda x:x.split(",")))
list_item_name
```
