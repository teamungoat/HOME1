```python
import pandas as pd
import numpy as np
```

```python
#1. df1과 df2에 대해 merge 연습, on, left_on, right_on, how

df1 = pd.DataFrame({"key": ["b", "b", "a", "c", "a", "a", "b"],
                    "data1": pd.Series(range(7), dtype="Int64")})
df2 = pd.DataFrame({"key": ["a", "b", "d"],
                    "data2": pd.Series(range(3), dtype="Int64")})
df1
df2
```

```python
pd.merge(df1,df2)
pd.merge(df1,df2, on='key')
pd.merge(df1,df2, how='outer')
pd.merge(df1,df2, how='left')
pd.merge(df1,df2, how='right')
pd.merge(df1,df2, left_on='key', right_on='key')
pd.merge(df1,df2, on='key')
pd.merge(df1,df2, left_on='data1', right_on='data2')
```

```python
##2. left과 right에 대해 merge 연습, on, left_on, right_on, how
left = pd.DataFrame({"key1": ["foo", "foo", "bar"],
                     "key2": ["one", "two", "one"],
                     "lval": pd.Series([1, 2, 3], dtype='Int64')})
right = pd.DataFrame({"key1": ["foo", "foo", "bar", "bar"],
                      "key2": ["one", "one", "one", "two"],
                      "rval": pd.Series([4, 5, 6, 7], dtype='Int64')})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=["key1", "key2"], how="outer")
pd.merge(left, right)
pd.merge(left, right, how='outer')
pd.merge(left, right, left_on=["key1", "key2"], right_on=["key1", "key2"])
pd.merge(left, right, left_on=["key1", "key2"], right_on=["key1", "key2"], how='outer')
pd.merge(left, right, on='key1')
pd.merge(left, right, left_index=True, right_index=True,how='outer')
```

```python
#3. merge 열 이름과 행 인덱스를 기준으로 병합 연습
left1 = pd.DataFrame({"key": ["a", "b", "a", "a", "b", "c"],
                      "value": pd.Series(range(6), dtype="Int64")})
right1 = pd.DataFrame({"group_val": [3.5, 7]}, index=["a", "b"])
left1
```

```python
pd.merge(left1,right1, left_on='key', right_index=True)
```

```python
#4. merge연습, index 기준
left2 = pd.DataFrame([[1., 2.], [3., 4.], [5., 6.]],
                     index=["a", "c", "e"],
                     columns=["Ohio", "Nevada"]).astype("Int64")
right2 = pd.DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]],
                      index=["b", "c", "d", "e"],
                      columns=["Missouri", "Alabama"]).astype("Int64")
left2, right2
```

```python
pd.merge(left2,right2, left_index=True, right_index=True, how='outer')
```

```python
#5. concat 연습, axis, join 속성 값 변경 설정
s1 = pd.Series([0, 1], index=["a", "b"], dtype="Int64")
s2 = pd.Series([2, 3, 4], index=["c", "d", "e"], dtype="Int64")
s3 = pd.Series([5, 6], index=["f", "g"], dtype="Int64")
s1, s2, s2
```

```python
pd.concat([s1,s2,s3])
pd.concat([s1,s2,s3], axis=1)
s1_s2=pd.concat([s1,s2], axis=1)
s3.name = 's3'
s1_s2.join(s3, how='outer')

```

```python
#6. stack, unstack 연습
df1 = pd.DataFrame(np.arange(6).reshape(3, 2), index=["a", "b", "c"],
                   columns=["one", "two"])
df2 = pd.DataFrame(5 + np.arange(4).reshape(2, 2), index=["a", "c"],
                   columns=["three", "four"])
df1,df2
```

```python
merged_df=pd.merge(df1,df2,right_index=True,left_index=True,how='outer')
merged_df.stack().unstack()
pd.concat([df1, df2]) # 미중복은 대각으로 접붙임
```

```python
pd.concat([df1.stack(), df2.stack()], axis=0) # 행으로 접붙이기
```

```python
pd.concat([df1.stack(), df2.stack()], axis=1) 
```
