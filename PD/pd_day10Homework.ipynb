```python
import pandas as pd
import numpy as np
```

```python
pieces=[]
for y in range(1880,2011):
    frame=pd.read_csv(f'babynames/yob{y}.txt', names=['name','sex','birth'])
    frame['year']=y
    pieces.append(frame)
```

```python
pieces
```

```python
names=pd.concat(pieces,ignore_index=True)
```

```python
names
```

```python
def get_rank(grp):
    grp['rank']=grp['birth'].rank()
    grp['count']=grp['birth'].count()
    return grp.sort_values('rank', ascending=False)
ranked_name=names.groupby('year').apply(get_rank)
name06_10=ranked_name.loc[2006:2010]
name06_10
```
