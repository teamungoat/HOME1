```python
def top3_petal_length(df):
    return df.sort_values(by="petal_length", ascending=False)[:3]

iris.groupby(iris.species).apply(top3_petal_length)[['petal_length']]
```

```python
pd.DataFrame(iris.groupby('species'))
```

```python
iris.groupby(iris.species).apply(lambda x: x.sort_values(by='petal_length').head(3))
```

```python
def myTop3(df):
    print(df)
    print("="*50)
iris.groupby(iris.species).apply(myTop3)
iris.groupby(iris.species).apply(myTop3)
```

```python
def myTop3(df):
    print(df.sort_values(by="petal_length",ascending=False))
    #print(df)
    print("="*50)
iris.groupby(iris.species).apply(myTop3)
```

```python
def q3(s):
    return pd.cut(s, 3, labels=['소','중','대'])

iris['p_l_class']=iris.groupby(iris.species).petal_length.transform(q3)
iris[['petal_length','p_l_class']]
```
